```python
# your code here

orders2 = orders[['CustomerID','Country','amount_spent']]
orders2.head(5)

```

```python
orders2.shape
```

```python
orders3 = orders2.groupby(['CustomerID'])['amount_spent'].agg(['sum'])
orders3
```

```python
VIP75 = np.percentile(orders3['sum'],75)
pre95 = np.percentile(orders3['sum'],95)

print(VIP75)
print(pre95)
```

```python
#cutoffs=[(orders3['sum']>=pre95),(orders3['sum']>=VIP75)]
cutoffs = [0, 1661.64, 5840.189199999983,99999999999999]

sum_labels = ['Ordinary','Preferred', 'VIP']
```

```python
orders3['Type of customer'] = pd.cut(orders3['sum'],cutoffs, labels=sum_labels)
orders3.head(10)

```

```python
orders4 =  orders3.merge(orders2, how='left', left_index=True, right_on='CustomerID')
orders4
```

```python
VIPcountry=orders4[orders4['Type of customer']=='VIP'].groupby(['Country'])['CustomerID'].agg(['nunique'])
VIPcountry.sort_values('nunique',ascending=False)
```

# Q3: How to identify which country has the most VIP+Preferred Customers combined?

Provide your solution for Q2 below:

```python

VIPcountry=orders4[(orders4['Type of customer']=='VIP')|(orders4['Type of customer']=='Preferred') ].groupby(['Country'])['CustomerID'].agg(['nunique'])
VIPcountry.sort_values('nunique',ascending=False)
```
